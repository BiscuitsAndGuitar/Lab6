    |      id      |  minutes  |
    |--------------------------|

    |    364322    |    150    |
    |    26177     |    20     |

Обратите внимание, что ширина столбцов заранее неизвестна и должна рассчитываться динамически, в зависимости от тех данных, которые были выбраны. 

````python
import pandas as pd
import random
````

````python
recipes = pd.read_csv('data/recipes_sample.csv')
random_recipes = recipes.sample(n=5)
recipes_string = random_recipes[['id', 'minutes']].to_string(index=False, header=False)
max_len = max(map(len, recipes_string.split() + ['id', 'minutes'])) + 4
headers = ['id', 'minutes']
print(f'|{headers[0]: ^{max_len}}|{headers[1]: ^{max_len}}|')
print('|', '-' * (2 * max_len + 1), '|', sep='')
for recipe in recipes_string.split('\n'):
    recipe = recipe.split()
    print(f'|{recipe[0]: ^{max_len}}|{recipe[1]: ^{max_len}}|')
````

````python
import requests
from bs4 import BeautifulSoup
````

````python
with open('data/steps_sample.xml') as xml_file:
    steps_sample = BeautifulSoup(xml_file, 'xml')
````

````python
steps_dict = {}

for recipe in steps_sample.find_all('recipe'):
    recipe_id = recipe.find('id').text.strip()
    steps_list = []
    
    for i, step in enumerate(recipe.find_all('step')):
        if i >= 2:  
            break
        steps_list.append(step.text.strip())
    
    steps_dict[recipe_id] = steps_list
````

````python
steps_dict
````

````python
def show_info(recipe_id):
    return f'''
    {recipes[recipes['id'] == recipe_id]['name'].values[0]}
    1. {steps_dict[str(recipe_id)][0]}
    2. {steps_dict[str(recipe_id)][1]}
    ------------------
    Автор: {recipes[recipes['id'] == recipe_id]['contributor_id'].values[0]}
    Среднее время приготовления: {recipes[recipes['id'] == recipe_id]['minutes'].values[0]}
    '''
````

````python
print(show_info(170895))
````
